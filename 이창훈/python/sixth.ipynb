```python
db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = mysql_user_info['user_id'],
    passwd = mysql_user_info['password'],
    db = 'pydb'
)

cursor = db.cursor()

sql = """
    create table power_data(
        no integer not null auto_increment primary key,
        voltage float not null,
        current float not null,
        vendor text null,
        name text not null,
        reg_date timestamp not null default now()
    )
"""

cursor.execute(sql)

db.commit()
db.close()

print('float 형태의 전력 데이터 관리 테이블 생성 성공!')
```

```python
import csv
```

```python
f = open('mysql2csv.csv', 'r', encoding='UTF-8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)
```

```python
cnt = 0

db = pymysql.connect(
    host = '127.0.0.1',
    port = 3306,
    user = mysql_user_info['user_id'],
    passwd = mysql_user_info['password'],
    db = 'pydb'
)

cursor = db.cursor()

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
        
    #print(row)
    sample_no = (row[0])
    voltage = (row[1])
    current = (row[2])
    vendor = (row[3])
    product_name = (row[4])
    reg_date = (row[5])
    
    #print(sample_no, voltage, current, vendor, product_name, reg_date)
    
    sql = """
        insert into power_data (no, voltage, current, vendor, name, reg_date) values (%s, %s, %s, %s, %s, %s)
    """
    
    cursor.execute(sql, (sample_no, voltage, current, vendor, product_name, reg_date))

db.commit()

f.close()
db.close()
```

```python
import json
import datetime
import os

# jupyter notebook 구동되는 위치에 kakao_msg 디렉토리 만들어야함
KAKAO_TOKEN_SAVED_FILE = "kakao_msg/kakao_token.json"

def save_tokens(file, token):
    with open(file, "w") as fp:
        json.dump(token, fp)
        
def load_tokens(file):
    with open(file) as fp:
        token = json.load(fp)
        
    return token

save_tokens(KAKAO_TOKEN_SAVED_FILE, token)
```
